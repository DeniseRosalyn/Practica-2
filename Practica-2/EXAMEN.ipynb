```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as ptl
import seaborn as sns
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
import requests
```

```python
url="https://news.google.com/covid19/map?hl=es-419&gl=PE&ceid=PE%3Aes-419"
uclient=urlopen(url)
pagehtml=uclient.read()
uclient.close()
print(pagehtml)
```

```python
pagesoup=soup(pagehtml, "html.parser")
page=pagesoup.find("div",{"class":"sOh CrmLxe"})
paises=page.findAll("tr",{"class":"sgXwHf wdLSAe YvL7re "})
```

```python
pais=[]
total_casos=[]
casos_nuevos_por_dia=[]
casos_por_millon_personas=[]
decesos=[]
link_grafico=[]

for i in paises:
  pais.append(i.th.div.text)
  total_casos.append(i.findAll("td")[0].text)
  casos_nuevos_por_dia.append(i.findAll("td")[1].text)
  casos_por_millon_personas.append(i.findAll("td")[3].text)
  decesos.append(i.findAll("td")[4].text)
  link_grafico.append(i.findAll("td")[2].img["src"])

```

```python
###Data frame 
lugar=list(range(1,41))
df=pd.DataFrame({"Pais":pais,"Total de casos":total_casos,"Casos nuevos por dia":casos_nuevos_por_dia,
                 "Casos por millon de personas":casos_por_millon_personas,"Gráfico últimos 60 dias":link_grafico,"Decesos":decesos},lugar)
print(df)
```

```python
#Convirtiendolo a csv
df.to_csv("estadisticas.csv")
```
